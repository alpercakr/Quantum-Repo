```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Assignment of the constants:
h = 6.62607015e-34 # Planck' constant
kb = 1.38e-23 # Boltzmann' constant
c = 3e8
```

```python
# temperature values
temps = np.array([3000, 4000, 5000])
nu = np.linspace(0, 1e15, 100)
```

```python
# Function to calculate the spectral radiance
def B(nu, T):
    x = (2*h*nu**3/c**2)*1/(np.exp(h*nu/(kb*T)) - 1)
    return x
```

```python
plt.figure(figsize=(10,7))
plt.title("Black Body Radiation")
t_values = ['T1: 3000', 'T2: 4000', 'T3: 5000']

for i in range(len(temps)):
    plt.plot(B(nu, temps[i]), label=t_values[i])
plt.legend(title='Temperature values')
plt.xlabel('Frequency')
plt.ylabel('Spectral radiance')
```
